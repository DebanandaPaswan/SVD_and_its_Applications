```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
#read image in grayscale
img = cv2.imread("images/beach.jpg", 0)
```

```python
img
```

```python
img.shape
```

```python
# to view the image
cv2.imshow('image', img)
```

```python
#obtain SVD
U, S, V = np.linalg.svd(img)
```

```python
# inspect shapes of the matrices
print(U.shape, S.shape, V.shape)
```

```python
#plot images with different components
comps = [638, 500, 400, 300, 200, 100]
```

```python
plt.figure(figsize=(16, 8))
```

```python
for i in range(6):
    low_rank = U[:,:comps[i]]@np.diag(S[:comps[i]])@V[:comps[i],:]
    if i == 0:
        plt.subplot(3,2,i+1), plt.imshow(low_rank, cmap='gray'), plt.axis('off'), plt.title("original image n_components ="+str(comps[i]) )
    else:
        plt.subplot(3,2,i+1), plt.imshow(low_rank, cmap='gray'), plt.axis('off'), plt.title("n_components ="+str(comps[i]) )
```
